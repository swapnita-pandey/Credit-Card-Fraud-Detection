Dealing with unbalanced data

Under-Sampling: Build a sample dataset from the original dataset containing similar distribution of legit transactions and fradulent transactions.

Since we know that the number of fradulent transactions is 492.

So, out of 284315 legit transactions we will take random 492 transactions and then join it with the fradulent transactions.

So, the final dataset will have equal number of legit and fradulent transactions.

This dataset will have uniform distribution of legit and fradulent datasets. This will help in doing better predictions using our machine learning model.

```python
legit_sample = legit.sample(n = 492)
```

Concatenating two Dataframes

If axis is 0 dataset is added one by one row wise.

Axis 0 - row

Axis 1 - column

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into features and targets.

X - features

Y - labels

```python
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and testing data

We are craeting 4 arrays X_train, X_test, Y_train, Y_test

test_size = 0.2 indicates that 20% of data will be used for testing and 80% of data will will be used for training.

We want to startify the data based on Y.

In Y the distributionof 0 and 1 can be very different in training data and testing data, using startify there will be evenly distributed 2 classes in both x_train and X_test.

random_state is just to randomize dataset in different ways.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify = Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

We had 984 values in our original dataset, about 80% that is 787 values go to X_train and 20% that is about 197 values go to Y_train.

Model Training

We use Logistic Regression for Binary Classification problems.

We are loading a instance of Logistic Regression model in a variable model.

```python
model = LogisticRegression()
```

Training the Logistic Regression Model with training data.

model.fit is used to fit our data according to the Logistic Regression Model.

```python
model.fit(X_train, Y_train)
```

Model Evaluation

Checking the performance of your model based on the accuracy score

Accuracy Score

Accuracy Score on Training data

We will predict the labels of the training data and store the predicted labels in X_train_prediction variable.

Then, we will compare the values predicted by our model (stored in X_train_prediction) with the original label values (stored in Y_train)

It will compare the values and give us the accuracy score and the accuracy score will be stored in training_data_accuracy variable.

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Training data accuracy: ', training_data_accuracy)
```

Accuracy Score on Testing data

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Testing data accuracy: ', testing_data_accuracy)
```

If the model is overfitted or under-fitted, the training and testing accuracy have a huge difference.

If the testing data accuracy is too large as compared to testing data accuracy, the model is over-fitted.

If the testing data accuracy is much less as compared to traing data accuracy, the model is under-fitted.
